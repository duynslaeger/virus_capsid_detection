```python
from skimage.io import imread, imshow, imsave
from skimage.color import gray2rgb
from matplotlib import pyplot as plt
import glob
from os import listdir
from pathlib import Path
```

```python
images = Path("Data/dataset/").glob('*.tif')
for image in images:
    print(image)
    im = imread(image)
    new = gray2rgb(im)
    plt.imshow(new)
    name = str(image)[13:20]
    print(name)
    imsave("dataset_gray2rgb/"+name+".jpg",new)
```

```python
for image in listdir("Data/dataset/"):
    path = 'Data/dataset/' + str(image)
    print(path)
    im = imread(path)
    new = gray2rgb(im)
    plt.imshow(new)
    imsave("dataset_gray2rgb/"+str(image)+".jpg",new)
```
